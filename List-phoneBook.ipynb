```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('finalized-v2.csv')

# Extract unique IDs and their allowing names from 'From' and 'Source' columns
id_table = df[['From', 'source']].drop_duplicates().reset_index(drop=True)

# Fill the 'Target' column using the 'id_table' DataFrame
df['target'] = df['To'].apply(lambda x: id_table.loc[id_table['From'] == x, 'source'].values[0] if x in id_table['From'].values else 0)

# Print the updated DataFrame
df.to_csv("modified_file.csv", index=False)
```
